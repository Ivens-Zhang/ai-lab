### 我们有城市空气质量数据，请做数据处理，绘制北京的PM2.5随时间的变化情况
####  数据来源： https://www.kaggle.com/uciml/pm25-data-for-five-chinese-cities


```python
import pandas as pd
from matplotlib import pyplot as plt

# 获取 csv 文件
file_path = "./data/北京每年1-12月的PM2.5结果.csv"
df = pd.read_csv(file_path)

# 将文件中数据按照 year 分组
data1 = df.groupby(by="year").sum()["PM_ave"]

_x = data1.index
_y = data1.values

#画图
plt.figure(figsize=(20,8),dpi=180)

plt.bar(range(len(_x)),_y, label='bj')

plt.xticks(range(len(_x)),_x)

plt.show()
```
